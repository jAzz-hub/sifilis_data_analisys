#### Substituindo dados sobre raças e Zona de Residência

```python
# Mapeamento de valores para as etnias
mapeamento_raca = {
    1: 'Branca',
    2: 'Preta',
    3: 'Amarela',
    4: 'Parda',
    5: 'Indígena',
    9: 'Não Assinalado na Ficha',
    "Em Branco": 'Em Branco'
}

mapeamento_resid = {
    1 :'Urbana',
    2 :'Rural',
    3 :'Periurbana',
    9 :'Ignorado',
    "Em Branco": 'Em Branco'
}

mapeamento_escolaridade = {
    1:'1ª a 4ª série incompleta do EF',
    2:'4ª série completa do EF',
    3:'5ª à 8ª série incompleta do EF',
    4:'Fundamental completo',
    5:'Médio incompleto ',
    6:'Médio completo ',
    7:'Superior incompleta',
    8:'Educação superior completa',
    9:'Ignorado',
    10: 'Não se aplica',
    "Em Branco": 'Em Branco'
}

mapeamento_trat = {
1 :"Penicilina G benzantina 2.400.000 UI ",
2 :"Penicilina G benzantina 4.800.000 UI ",
3 :"Penicilina G benzantina 7.200.000 UI",
4 :"Outro esquema ",
5 :"Não realizado ",
9 :"Ignorado",
"Em Branco": 'Em Branco'
}

mapeamento_trigest = {
1 :"1º",
2 :"2º",
3 :"3º",
4 :"Idade Gestacional Ignorada",
9 :"Ignorado",
"Em Branco": 'Em Branco',
}

# Substituir os valores na coluna 'etnia'
df['Raca'] = df['Raca'].map(mapeamento_raca)
df['Residencia'] = df['Residencia'].map(mapeamento_resid)
df['Escolaridade'] = df['Escolaridade'].map(mapeamento_escolaridade)
df['Tratamento'] = df['Tratamento'].map(mapeamento_trat)
df['TriGestacional'] = df['TriGestacional'].map(mapeamento_trigest)
```

```python


# Função para converter a formatação
def formatar_penicilina(valor):
    if isna(valor):  # Se for NaN, mantém como está
        return valor
    elif "4.800.000" in valor:
        return "PGB 4.800.000 UI"
        print("AQUI")
    elif "2.400.000" in valor:
        return "PGB 2.400.000 UI"
        print("AQUI")
    elif "7.200.000" in valor:
        return "PGB 7.200.000 UI"
        print("AQUI")
    else:
        return valor  # Retorna o valor original caso não se encaixe nos padrões
        print("AQUI")

# Aplicando a conversão na coluna de tratamento
df['Tratamento'] = df['Tratamento'].apply(formatar_penicilina)

# Exibindo alguns valores para conferir
print(df['Tratamento'].unique())
```
